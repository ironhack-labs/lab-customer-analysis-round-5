```python
#grouping response per month

group1 = analysis.groupby('month')['response'].value_counts(normalize=True).unstack().fillna(0)
group1.reset_index(inplace=True)
group1['response_rate'] = group1['Yes'] * 100
group1
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='response_rate', data=group1, palette="Reds")
```
