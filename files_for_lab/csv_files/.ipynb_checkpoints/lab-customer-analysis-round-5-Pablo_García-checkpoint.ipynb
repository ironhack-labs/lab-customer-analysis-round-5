```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
# Define constants that will be used during processing:
CSV_FILE = 'numerical_from_lab4.csv'
Y_VAR = 'total_claim_amount'
THRESHOLD = 3
COLS_TO_LOG = ['customer_lifetime_value', 'income', 'monthly_premium_auto']
COLS_TO_BOX = ['months_since_last_claim', 'months_since_policy_inception',
       'number_of_open_complaints', 'number_of_policies',
       'month']
TEST_SIZE = 0.4
RANDOM_STATE = 100
```

```python
# Define the functions that will be used during processing

def import_to_df(csv_file):
    df = pd.read_csv(csv_file)
    df = df.drop(['Unnamed: 0'], axis = 1)
    return df

def split_data(df, y_var):
    x = df.drop([y_var], axis=1)
    y = df[y_var]
    return x,y

def plot_hist(df):
    num_cols = len(df.columns)
    num_subplots = num_cols // 2 + num_cols % 2
    fig, axis = plt.subplots(2,num_subplots, figsize=(12,8))  
    axis = axis.ravel()
    for i, col in enumerate(df.columns):
        sns.histplot(df[col], kde=True, ax = axis[i]).set_xlabel(col)     
    plt.tight_layout()
    plt.show()
    
def plot_box(df):
    num_cols = len(df.columns)
    num_subplots = num_cols // 2 + num_cols % 2
    fig, axis = plt.subplots(2,num_subplots, figsize=(12,8))  
    axis = axis.ravel()
    for i, col in enumerate(df.columns):
        sns.boxplot(x = df[col], ax = axis[i]).set_xlabel(col)     
    plt.tight_layout()
    plt.show()
    
# def filter_outliers(df, threshold):
#     mask = pd.Series(data=True, index=df.index)
#     for col in df.columns:
#         q1 = df[col].quantile(0.25)
#         q3 = df[col].quantile(0.75)
#         iqr = q3 - q1
#         lower_limit = q1 - (threshold * iqr)
#         upper_limit = q3 + (threshold * iqr)
#         col_mask = (df[col] >= lower_limit) & (df[col] <= upper_limit)
#         mask = mask & col_mask
#     return df[mask]

def replace_inf(i):
    if np.isfinite(i):
        return i
    else:
        return np.NAN
    
def log_transformer(df, cols_to_log):
    df_copy = df.copy()
    for col in cols_to_log:
        df_copy[col] = np.log(df_copy[col])
        df_copy[col] = list(map(replace_inf, df_copy[col]))
        df_copy[col] = df_copy[col].fillna(df_copy[col].mean())
    return df_copy

def box_cox_transformer(df, cols_to_transform):
    df_copy = df.copy()
    for col in cols_to_transform:
        df_copy[col] = np.where(df_copy[col] <= 0, 0, df_copy[col])
        df_copy[col].replace(0,df[col].mean(), inplace=True)
        transformed_col, _ = stats.boxcox(df_copy[col])
        df_copy[col] = transformed_col
    return df_copy

def min_max_scaler(df):
    df_copy = pd.DataFrame(MinMaxScaler().fit(df).transform(df))
    return df_copy

def standard_scaler(df):
    columns = df.columns
    df_copy = pd.DataFrame(StandardScaler().fit(df).transform(df))
    df_copy.columns = columns
    return df_copy

def train_test_split(df_x, df_y, test_size, random_state):
    X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=test_size,random_state=random_state)
    train_set = pd.concat([X_train, y_train], axis=1)
    test_set = pd.concat([X_test, y_test], axis=1)
    return X_train, y_train
```

```python
#1. Import the data:
df = import_to_df(CSV_FILE)
df.shape
```

```python
#2. Plot data 

plot_hist(df)
```

```python
#3. Split X and Y variables

df_x, df_y = split_data(df,Y_VAR)
df_x.head()
df_y.head()
```

```python
#5. Use Log transformation to normalise data
df_x_log = log_transformer(df_x, COLS_TO_LOG)
```

```python
plot_hist(df_x_log)
```

```python
#6. Use Box-Cox Transformation to normalise data
df_x_cox = box_cox_transformer(df_x, ['customer_lifetime_value', 'income', 'monthly_premium_auto',
       'months_since_last_claim', 'months_since_policy_inception','number_of_policies', 'month'])
```

```python
plot_hist(df_x_cox)
```

```python
#7. Standardise data using Standard Scaler:
df_x_cox_std = standard_scaler(df_x_cox)
```

```python
plot_hist(df_x_cox_std)
```

```python
#8.Drop columns that wont be included in the model
df_x = df_x_cox_std.drop(['number_of_open_complaints', 'number_of_policies', 'month'], axis=1)
df_x
```
